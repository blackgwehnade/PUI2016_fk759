```python
import os as os
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
%pylab inline
```

```python
dfx = pd.read_csv('Citibike_Member_Ridership_data.csv')
dfx.drop(dfx.columns[[1, 2, 3, 4, 5, 7, 8]], axis=1, inplace=True)
dfx.head(7)
```

```python
dfx['Date'] = pd.to_datetime(dfx['Date'])
dfx.head()
```

```python
plt.figure(figsize=(16, 4))
plt.title('Citibike 24-Hour Passes as a function of Date')
plt.xlabel('Date')
plt.ylabel('24-Hour Passes Purchased from midnight to 11:59pm')
plt.plot(np.array(dfx['Date']), np.array(dfx['24-Hour Passes Purchased (midnight to 11:59 pm)']))
plt.bar(np.array(dfx['Date']), np.array(dfx['24-Hour Passes Purchased (midnight to 11:59 pm)']))
plt.show()
```
